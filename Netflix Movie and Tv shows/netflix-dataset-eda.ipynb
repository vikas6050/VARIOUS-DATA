```python
import pandas as pd
pd.options.mode.chained_assignment = None
import numpy as np
import plotly.express as px
import matplotlib
```

```python
df=pd.read_csv('/kaggle/input/netflix-shows/netflix_titles.csv')
df.head()
```

```python
df.info()
```

```python
df['date_added'] = pd.to_datetime(df["date_added"].str.strip(), format="%B %d, %Y")
```

```python
df.describe()
```

Exploring 10 oldest movies in the dataset

```python
df.sort_values(by=['release_year'],ascending=False).tail(10)
```

creating a histogram by released year

```python
px.histogram(df, x="release_year").show()
```

creating a histogram by date added

```python
px.histogram(df, x="date_added").show()
```

extracting month and day from date added column

```python
df["date_added_month"] = df["date_added"].dt.month.fillna('unknown')
df["date_added_day"] = df["date_added"].dt.day.fillna('unknown')
```

```python
px.histogram(df, x="date_added_month").show()
```

Most content is added in July and least in February

```python
px.histogram(df, x="date_added_day").show()
```

Most content is added on the 1st day of the month and then 15th

```python
country_df = df["country"].str.split(',', expand=True)
df_copy = df.copy()
country_df = pd.concat([country_df, df_copy], axis=1)
country_df
```

```python
country_df = country_df.melt(id_vars=['type', 'title'], value_vars=range(12), value_name= 'Country')
country_df
```

```python
country_df = country_df[country_df['Country'].notna()]
country_df['Country'] = country_df['Country'].str.strip()
country_df.drop(['variable'], axis=1)
```

```python
top_country = country_df.groupby(['Country']).size().to_frame().sort_values([0], ascending = False).head(10).reset_index()
top_country.rename(columns={0: "Count"}, inplace=True)
px.histogram(data_frame=top_country, x="Country", y='Count').update_xaxes(categoryorder='total descending').show()
```

```python
cast_df = df['cast'].str.split(',', expand=True)
cast_df = cast_df.melt(value_name = 'Actor')
cast_df = cast_df[cast_df['Actor'].notna()]
```

```python
cast_df
```

```python
top_cast = cast_df.groupby(['Actor']).size().to_frame().sort_values([0], ascending = False).head(15).reset_index()
top_cast.rename(columns={0: "Count"}, inplace=True)
px.histogram(top_cast, x="Actor", y = "Count", title='Top 15 actors').update_xaxes(categoryorder='total descending').show()
```

```python
type_df=df.groupby(['type']).size().to_frame().reset_index()
type_df.rename(columns={0: "Count"}, inplace=True)
px.pie(type_df, values='Count', names='type')
```

Since all top 15 actors are Indian, unfilter India to see the top actors from other countries

```python
no_ind=df.loc[df['country'].str.contains('India')==False]
no_ind
```

```python
cast_df_2 = no_ind['cast'].str.split(',', expand=True)
cast_df_2 = cast_df_2.melt(value_name = 'Actor')
cast_df_2 = cast_df_2[cast_df_2['Actor'].notna()]
```

```python
top_cast_2 = cast_df_2.groupby(['Actor']).size().to_frame().sort_values([0], ascending = False).head(15).reset_index()
top_cast_2.rename(columns={0: "Count"}, inplace=True)
px.histogram(top_cast_2, x="Actor", y = "Count", title='Top 15 actors (Non-Indian)').update_xaxes(categoryorder='total descending').show()
```
